```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
k = 8.99e9  # Coulomb constant in N·m²/C²
q1 = 1e-6   # Charge 1 in Coulombs
q2 = -1e-6  # Charge 2 in Coulombs


x = np.linspace(-0.1, 0.1, 200)
y = np.linspace(-0.1, 0.1, 200)
X, Y = np.meshgrid(x, y)

# Positions of charges :Place +q and –q symmetrically about origin
x1, y1 = -0.03, 0
x2, y2 = 0.03, 0

#Euclidean Distance from each grid point to each charge
r1 = np.sqrt((X - x1)**2 + (Y - y1)**2)
r2 = np.sqrt((X - x2)**2 + (Y - y2)**2)

# Prevent divide by zero (replace 0 with a small number)
r1[r1 == 0] = 1e-12
r2[r2 == 0] = 1e-12

# Electric potential calculation
V = k * (q1 / r1 + q2 / r2)

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, V, cmap='plasma', edgecolor='none')

ax.set_title('Electric Potential of Two Point Charges')
```
